```python
import nltk
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
line_len = dict()
with open("/home/work/Data/raw.para/judicial.zh.split") as inputfile:
    for line in inputfile:
         if line != "\n" and line != "":
                line = line.split(" ")
                if len(line) not in line_len:
                    line_len[len(line)] = 1
                elif len(line) in line_len:
                    line_len[len(line)] += 1
            
plt.bar(range(len(line_len)), line_len.values(), align='center')
plt.xticks(range(len(line_len)), line_len.keys())

plt.show()
sortd = sorted(line_len.items(), key=lambda x: x[1],reverse=True)

x_val = [x[0] for x in sortd[:10]]
y_val = [x[1] for x in sortd[:10]]

plt.plot(x_val,y_val)
plt.plot(x_val,y_val,'or')
plt.xlabel('sentence Length')
plt.ylabel('Frequency')
plt.show()
```

```python
en_line_len = dict()
with open("/home/work/Data/raw.para/judicial.en.split") as inputfile:
    for line in inputfile:
        if line != "\n" && line != "":
            line = nltk.word_tokenize(line.decode("utf-8"))
            if len(line) not in en_line_len:
                en_line_len[len(line)] = 1
            elif len(line) in en_line_len:
                en_line_len[len(line)] += 1
sortde = sorted(en_line_len.items(), key=lambda x: x[1],reverse=True)

x_val = [x[0] for x in sortde[:10]]
y_val = [x[1] for x in sortde[:10]]

plt.plot(x_val,y_val)
plt.plot(x_val,y_val,'or')
plt.xlabel('sentence Length')
plt.ylabel('Frequency')
plt.show()
```
